```python
import geopandas as gpd

# Load shapefile
gdf = gpd.read_file("PMH Ouarzazat.shp")

# Check current CRS
print(gdf.crs)
```

Convert to WGS84

```python
gdf = gdf.to_crs(epsg=4326)

print(gdf.crs)
```

```python
print(gdf.total_bounds)
```

Export New WGS84 Shapefile

```python
gdf.to_file("PMH_Ouarzazat_WGS84.shp")
```
